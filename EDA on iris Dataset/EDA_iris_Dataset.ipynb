```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
import numpy as np
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("iris.csv")
print(df.shape)
```

```python
df
```

```python
print(df.columns)
```

```python
print(df["species"].value_counts())
```

```python
df.plot(kind="scatter", x = "sepal_length", y="sepal_width")
plt.show()
```

```python
sbn.set_style("whitegrid")
sbn.FacetGrid(df, hue = "species", size=4) \
    .map(plt.scatter, "sepal_length", "sepal_width") \
    .add_legend()
plt.show()
```

```python
sbn.set_style("whitegrid")
pal = ["blue", "green", "yellow"]
sbn.pairplot(df, hue= "species" ,palette= pal, size= 3)
plt.show()
```

```python
pal1 = ["blue","green","orange"]
sbn.FacetGrid(df, hue = "species",palette = pal1 ,size= 5)\
    .map(sbn.distplot, "petal_length") \
    .add_legend()
plt.show()
```

```python
sbn.FacetGrid(df, hue="species",palette = pal1, size= 5) \
    .map(sbn.distplot, "petal_width")\
    .add_legend()
plt.show()
```

```python
pal2 =["yellow","green","orange"]
sbn.FacetGrid(df, hue= "species", palette=pal2, size= 5 )\
.map(sbn.distplot, "sepal_length")\
.add_legend()
plt.show()
```

```python
sbn.FacetGrid(df, hue= "species", palette=pal2, size= 5 )\
.map(sbn.distplot, "sepal_width")\
.add_legend()
plt.show()
```

```python
df_setosa = df.loc[df["species"] == "setosa"];
df_virginica = df.loc[df["species"] == "virginica"];
df_versicolor = df.loc[df["species"] == "versicolor"]
```

```python
print(np.mean(df_setosa["petal_length"]))
print(np.mean(df_virginica["petal_length"]))
print(np.mean(df_versicolor["petal_length"]))
```

```python
print(np.std(df_setosa["petal_length"]))
print(np.std(df_virginica["petal_length"]))
print(np.std(df_versicolor["petal_length"]))
```

```python
print(np.percentile(df_setosa["petal_length"],np.arange(0,100,25)))
print(np.percentile(df_virginica["petal_length"],np.arange(0,100,25)))
print(np.percentile(df_versicolor["petal_length"],np.arange(0,100,25)))

```

```python
print(np.percentile(df_setosa["petal_length"],90))
```

```python
from statsmodels import robust
print(robust.mad(df_setosa["petal_length"]))
print(np.std(df_setosa["petal_length"]))
```
